```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your original DataFrame is named df
# Step 1: Clean BHK column
df['BHK'] = df['BHK'].astype(str).str.extract(r'(\d+)').astype(float)

# Step 2: Drop rows with missing BHK or Price
df_clean = df[['BHK', 'Price_in_cr']].dropna()

# Step 3: Convert BHK to int for cleaner labels
df_clean['BHK'] = df_clean['BHK'].astype(int)

# Step 4: Sort BHK values
sorted_bhk = sorted(df_clean['BHK'].unique())

# Step 5: Create the plot
plt.figure(figsize=(8,6))
ax = sns.barplot(
    x="BHK",
    y="Price_in_cr",
    data=df_clean,
    estimator="mean",
    errorbar=None,
    palette="Set2",
    order=sorted_bhk
)

# Step 6: Add count labels above each bar
counts = df_clean['BHK'].value_counts().to_dict()
for p in ax.patches:
    bhk = int(p.get_x() + p.get_width()/2)
    count = counts.get(bhk, 0)
    height = p.get_height()
    ax.text(p.get_x() + p.get_width()/2, height + 0.05, f'N={count}', ha='center', fontsize=10)

# Step 7: Final touches
plt.title("Average Price by BHK with Listing Counts", fontsize=14)
plt.xlabel("Number of Bedrooms (BHK)")
plt.ylabel("Average Price (in Cr)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Filter for apartments
apartments = df[df['Property_Type'].str.lower() == 'apartment'].copy()

# Step 2: Ensure BHK is numeric
apartments['BHK'] = pd.to_numeric(apartments['BHK'], errors='coerce')

# Step 3: Define amenity columns
amenity_cols = ['Gym_status', 'Club_House_status', 'Lift_status', 'Swimming_Pool_status']

# Step 4: Create binary flags for amenity presence
for col in amenity_cols:
    apartments[col + '_present'] = apartments[col].astype(str).str.lower().apply(
        lambda x: 'no' not in x
    )

# Step 5: Count presence per BHK
presence_flags = [col + '_present' for col in amenity_cols]
amenity_counts = apartments.groupby('BHK')[presence_flags].sum()

# Step 6: Rename columns for clarity
amenity_counts.columns = ['Gym', 'Club House', 'Lift', 'Swimming Pool']

# Step 7: Drop BHKs with zero amenities
amenity_counts = amenity_counts.loc[amenity_counts.sum(axis=1) > 0]

# Step 8: Plot stacked bar chart (only if data exists)
if not amenity_counts.empty:
    amenity_counts.plot(kind='bar', stacked=True, figsize=(12,6),
                        color=['#d62728', '#1f77b4', '#ff9896', '#17becf'])

    plt.title('Amenity Distribution by BHK (Apartments Only)')
    plt.xlabel('BHK')
    plt.ylabel('Number of Properties with Amenity')
    plt.legend(title='Amenity')
    plt.xticks(rotation=0)
    plt.tight_layout()
    plt.show()
else:
    print("No apartment properties with amenities found for any BHK value.")
```
